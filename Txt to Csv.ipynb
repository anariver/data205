```python
# Creating 
txt_file_path = "xfinity_hotspots.txt"
txt_file_content = ""
with open(txt_file_path) as txt_file:
    txt_file_content = txt_file.read()
```

```python
txt_file_content = "\n".join(txt_file_content.split("Xfinity WiFi"))
print(txt_file_content.strip("\n"))
```

```python
txt_file_content = "\n".join(txt_file_content.split("Get directions to this location"))
print(txt_file_content)
```

```python
txt_file_content = "\n".join(txt_file_content.split("Search Results"))
print(txt_file_content)
```

```python
filter(None, txt_file_content.split("\n"))
# txt_file_content = "\n".join(txt_file_content.split("\n"))
# print(txt_file_content)
```

```python
filtered_data = list(filter(None, txt_file_content.split("\n")))
```

```python
csv_data = []
for i, j in enumerate(filtered_data):
    if i % 2 == 0:
        csv_data.append(j + ", " + filtered_data[i+ 1])
        print(i, j)
```

```python
with open("xfinity.csv", "w") as csv_out:
    csv_out.write("\n".join(csv_data))
```
