```python
# importing the packages time and beautiful soup
import time
from random import uniform

from bs4 import BeautifulSoup
import requests

# Creating a base url for the 14 pages of search results
base_url = "https://nces.ed.gov/ccd/schoolsearch/"

# Loading the main url that shows the results of all MCPS schools on NCES website
url = "https://nces.ed.gov/ccd/schoolsearch/school_list.asp?Search=1&County=Montgomery%20County&State=24&SchoolPageNum="

# Loop that go through the 14 pages of search results and get each link 
pages = []
for page_num in range(1, 15):
    print(url + str(page_num))
    r = requests.get(url + str(page_num))
    pages.append(r.text)
    time.sleep(uniform(1, 5))
```

```python
# Parsing each one of the 14 page results to obtain the links for each school
school_page_links = []
for page in pages:
    # calling the parser to go through each page, 
    page_soup = BeautifulSoup(page, "html.parser")
    school_td_soup = page_soup.find_all("td", attrs={"bgcolor":"#EDFFE8"})
    for i in school_td_soup:
        school_page_links.append(i.find("a")["href"])
```

```python
# Gong through each the pages and adding the "base url" in front of the scraped 209 links
school_pages = []
for school_page in school_page_links:
    print(base_url + school_page)
    r = requests.get(base_url + school_page)
    school_pages.append(r.text)
    time.sleep(uniform(1, 5))
```

```python
# Creating funtion that goes through each school ink and obtaining data for school name, enrollment, and number of FARMS and certified students
def get_school_data(source):

    school_soup = BeautifulSoup(source, "html.parser")
    data = {}

    school_name_soup = school_soup.find_all("td", attrs={"bgcolor":"#EDFFE8"})
    for i in school_name_soup:
        i = i.find("font", attrs={"size": "3"})
        if i:
            data["Name"] = i.text
            
    school_enrollment_soup = school_soup.find_all("td", attrs={"bgcolor":"#E8FCFF"})
    for i in school_enrollment_soup:
        i = i.find_all("font", attrs={"size": "3"})
        if i:
            data["Total Enrollment"] = i[1].text
            
    school_assistance_soup = school_soup.find_all("td", attrs={"bgcolor":"#FBE8E8"})
    for i in school_assistance_soup:
        i = i.find_all("font", attrs={"size": "2"})
        if len(i) == 3:
            for fonts in i:
                key, value = fonts.text.split(": ")
                data[key] = value
                
    return data
```

```python
# From parsed above, creating a csv file
mcps_demo_file = "mcps_demo.csv"

import pandas as pd

df = pd.DataFrame(columns=['Name', 'Total Enrollment', 'Free lunch eligible1', 'Reduced-price lunch eligible1', 'Directly certified2'])
for school_page in school_pages:
    row_data = get_school_data(school_page)
    df = df.append(row_data, ignore_index=True)

df.to_csv(mcps_demo_file)
```
